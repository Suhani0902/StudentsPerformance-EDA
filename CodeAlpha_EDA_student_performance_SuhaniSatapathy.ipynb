```python
# Upload the dataset
from google.colab import files
uploaded = files.upload()  # Choose StudentsPerformance.csv from your device

#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the dataset
df = pd.read_csv('StudentsPerformance.csv')
print("✅ File loaded successfully!\n")

#Basic Information
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nStatistical Summary:\n", df.describe(include='all'))

#Categorical and Numerical Columns
categorical_cols = df.select_dtypes(include='object').columns
numeric_cols = df.select_dtypes(exclude='object').columns

print("\nCategorical Columns:", list(categorical_cols))
print("Numeric Columns:", list(numeric_cols), "\n")

#Preview of dataset
print("First 5 rows of dataset:\n")
display(df.head())

#Data Distributions
plt.figure(figsize=(12,4))
for i, col in enumerate(numeric_cols):
    plt.subplot(1, len(numeric_cols), i+1)
    sns.histplot(df[col], kde=True, color='teal')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

#Average Scores by Gender
sns.barplot(data=df, x='gender', y='math score', estimator='mean', palette='pastel')
plt.title("Average Math Score by Gender")
plt.show()

#Writing Score by Parental Education
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='parental level of education', y='writing score', estimator='mean', palette='muted')
plt.xticks(rotation=45)
plt.title("Writing Score by Parental Education")
plt.show()

#Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

#Insights from Data
print("\nAverage Scores by Gender:\n", df.groupby('gender')[['math score','reading score','writing score']].mean(), "\n")
print("Average Scores by Lunch Type:\n", df.groupby('lunch')[['math score','reading score','writing score']].mean(), "\n")

#Quick Observation Summary
print("Observations:")
print("1. Reading and Writing scores are highly correlated.")
print("2. Female students tend to score higher in Reading and Writing, while males slightly lead in Math.")
print("3. Students with 'standard' lunch tend to have better overall performance.")
print("4. Higher parental education levels correspond to higher average student scores.")


```
